```python
import jax.numpy as jnp
import numpy as np
import matplotlib.pyplot as plt
from jax import vmap
```

```python
def sp(x,A,T):
    return A*jnp.exp(-x*x/T)
```

```python
xl=jnp.array([-1.0,1.0])
Al=jnp.array([1.0,1.5])
Tl=jnp.array([0.2,0.1])
```

```python
spv=vmap(sp,(0,None,None),0)
```

```python
x=jnp.linspace(-2.0,2.0,1000)
```

```python
plt.plot(x,spv(x,1.0,0.2))
```
